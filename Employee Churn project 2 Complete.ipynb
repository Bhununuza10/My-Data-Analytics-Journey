### Checking for relationships within the dataset(CORRELATION)

```python
# Checking for relationships within the dataset(CORRELATION)

# * Positive correlation means relationship between variables is moving in the same direction
# * Negative correlation means relationship between variables is moving in opposite direction
# * Zero correlation means there is no relationship between said variables(columns)
data.corr()
```

```python
# Visualisation of correlation between variables(columns)

lft = ["Department","number_project","time_spend_company","Work_accident","promotion_last_5years","salary"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.subplots(figsize=(18,15))

for p,q in enumerate(lft):
    plt.subplot(4,2,p+1)
    plt.subplots_adjust(hspace=1.0)
    sns.countplot(x=q,data=data, hue = "left")
    plt.xticks(rotation=90)
```

#### Analysis of visuals above:
######
     * Sales department experiences the most number of employees leaving the organization
     * Employees with 3 -4 projects are the ones that leave the organization the most
     * Most employees leave the company after spending 2- 4 years in the company, 3years being the most
     * Employees with 0 promotions in the last 5 years tend to be ones most likely to leave the company
     * high level of mmployees with low and medium salaries leaving the organization
